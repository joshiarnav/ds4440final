```python
import matplotlib.pyplot as plt

# Path to the text file containing the accuracies
file_path = 'results/training_accs.txt'

# Read the accuracies from the file
accuracies = []
with open(file_path, 'r') as file:
    for line in file:
        # Convert each line to a float and append to the list
        accuracies.append(float(line.strip()))

# Generate the x-axis values (every 20 iterations)
iterations = [i * 20 for i in range(len(accuracies))]

# Plotting the accuracies
plt.figure(figsize=(10, 5))
plt.plot(iterations, accuracies, marker='o', linestyle='-', color='b')
plt.title('Model Accuracy per Iteration')
plt.xlabel('Iteration')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()
```

```python
# Path to the text file containing the accuracies
file_path = 'results/training_accs_2.txt'

# Read the accuracies from the file
accuracies_2 = []
with open(file_path, 'r') as file:
    for line in file:
        # Convert each line to a float and append to the list
        accuracies_2.append(float(line.strip()))

# Generate the x-axis values (every 20 iterations)
iterations_2 = [1400 + (i * 20) for i in range(len(accuracies_2))]
```

```python
total_iterations = iterations + iterations_2
total_accuracies = accuracies + accuracies_2

# Plotting the accuracies
plt.figure(figsize=(10, 5))
plt.plot(total_iterations, total_accuracies, marker='o', linestyle='-', color='b')
plt.title('Model Accuracy per Iteration')
plt.xlabel('Iteration')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()
```

```python
# Plotting the accuracies
plt.figure(figsize=(10, 5))
for i, acc in enumerate(accuracies):
    if acc > 0.915:
        # Highlight accuracies greater than 0.915 with a red color and a larger marker
        plt.scatter(iterations[i], acc, color='red', s=100)  # s is the size of the marker
    else:
        # Normal accuracies in blue
        plt.scatter(iterations[i], acc, color='blue', s=50)

plt.plot(iterations, accuracies, marker='o', linestyle='-', color='b', label='Accuracy')
plt.title('Model Accuracy per Iteration')
plt.xlabel('Iteration')
plt.ylabel('Accuracy')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import json

# Load data from JSON file
with open('results/map_values.json', 'r') as file:
    data = json.load(file)

# Extract iterations, mAP, and accuracy values from the data
iterations = [item["iteration"] for item in data]
mAP_values = [item["mAP"] for item in data]
accuracy_values = [item["accuracy"] for item in data]

# Create the plot with both mAP and accuracy
plt.figure(figsize=(10, 6))
plt.plot(iterations, mAP_values, marker='o', linestyle='-', color='b', label='mAP')
plt.plot(iterations, accuracy_values, marker='s', linestyle='--', color='r', label='Accuracy')
plt.title('mAP and Accuracy by Iteration')
plt.xlabel('Iteration Number')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```
